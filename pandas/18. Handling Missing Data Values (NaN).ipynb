```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Frame from cars.csv
cereal = pd.read_csv('../datasets/cereal.csv')
cereal.head()
```

When we get a dataframe, it is always a good idea to check for any missing values

```python
cereal.isna()
```

If we chain `.isna()` with `.any()`, we get one value per variable, only showing for the column that might have NaN

```python
cereal.isna().any()
```

We can also take the sum to get the count of an NaN in each column

```python
cereal.isna().sum()
```

With the sum, we could even plot the missing values in the dataset

```python
cereal.isna().sum().plot(kind='bar')
plt.show()
```

So if we have missing values in the dataset, what can we do about them?
- We could **remove the whole row that have a missing value cell: `cereal.dropna()`**
- We could **replace NaN with a different value, like 0: `cereal.fillna(0)`
