```python
import pandas as pd
```

```python
# Data Frame from cars.csv
cars = pd.read_csv('../datasets/cars.csv', index_col=0)
cars.head()
```

```python
# Sorting by cars_per_cap
cars.sort_values(by=['cars_per_cap']) # Default is ascending
```

```python
# Sorting in a descending order
cars.sort_values(by=['cars_per_cap'], ascending=False)
```

```python
# Sorting using multiple columns
cars.sort_values(by=['cars_per_cap', 'drives_right'], ascending=[False, True])
```
